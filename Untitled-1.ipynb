```python
import os
from langchain_community.document_loaders import PyPDFLoader
from langchain.embeddings import OllamaEmbeddings
from langchain.vectorstores import Chroma

# Load and process the document
def process_document(pdf_path, persist_directory="db"):
    loader = PyPDFLoader(pdf_path)
    documents = loader.load()

    # Generate embeddings using LLaMA 3.1 (via Ollama)
    embeddings = OllamaEmbeddings(model="llama3.1")

    # Store embeddings in ChromaDB
    vectorstore = Chroma.from_documents(documents, embeddings, persist_directory=persist_directory)
    vectorstore.persist()

    print("Document processed and stored in vector database.")
```

```python
import os
from langchain_community.vectorstores import Chroma  # Updated import
from langchain.chains import RetrievalQA  # Updated import
from langchain.llms import Ollama  # Updated import for OllamaLLM
from langchain.embeddings import OllamaEmbeddings  # We already have this from previous cells
from langchain.prompts import PromptTemplate

# Load stored embeddings and vector store
persist_directory = "db"
vectorstore = Chroma(persist_directory=persist_directory, embedding_function=OllamaEmbeddings(model="llama3.1"))

# Load the LLaMA 3.1 model
llm = Ollama(model="llama3.1")

# Create a retrieval-based QA chain
qa_chain = RetrievalQA.from_chain_type(llm, retriever=vectorstore.as_retriever(), chain_type="stuff")

def chat_with_llama(user_query):
    if not user_query:
        return "Query is required"
    
    response = qa_chain.run(user_query)
    return response
```

```python
query = "Can you summarize GAMP 5 version 2?"
response = chat_with_llama(query)
print("LLaMA Response:", response)
```
