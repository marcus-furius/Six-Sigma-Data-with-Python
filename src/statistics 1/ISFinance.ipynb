```python
import os  # For working with file paths
import pandas as pd  # For data manipulation and loading
import matplotlib.pyplot as plt  # For creating plots
```

```python
data_file = r'C:\Users\dharris\OneDrive - Jazzpharma.com\Project\Jazz.Python\Six Sigma Data with Python\data\IS Budget Sheet.csv' # Full path to the CSV file
df = pd.read_csv(data_file, na_values="NA")
df
```

```python

# Drop the specified columns
columns_to_drop = list(df.columns[:4]) + list(df.columns[17:20]) + list(df.columns[21:61]) + list(df.columns[66:106]) + list(df.columns[113:])
df_cleaned = df.drop(columns=columns_to_drop)

# Display the resulting DataFrame
print(df_cleaned)
df_cleaned.to_csv(r'C:\Users\dharris\OneDrive - Jazzpharma.com\Project\Jazz.Python\Six Sigma Data with Python\data\IS Budget Sheet Cleaned.csv', index=False)
df_cleaned
```

```python
filtered_data = df_cleaned[
    (df_cleaned['Classification'] != 'Project') &
    (df_cleaned['Classification'] != '2025 Project') &
    (df_cleaned['CapEx / OpEx'] != 'Cap Ex')
]
filtered_data.head(10)    
```

```python
# Aggregating data by Department, Functional Area, Vendor, and Cost Type
# Convert cost-related columns to numeric for aggregation
for column in ['2024 Original Forecast', '2025 Original Forecast']:
    filtered_data[column] = (
        filtered_data[column]
        .replace('[\$,]', '', regex=True)
        .astype(float, errors='ignore')
    )

# Aggregating data by the chosen categories
department_summary = filtered_data.groupby('Department')[['2024 Original Forecast', '2025 Original Forecast']].sum()
functional_area_summary = filtered_data.groupby('Functional Area')[['2024 Original Forecast', '2025 Original Forecast']].sum()
vendor_summary = filtered_data.groupby('Vendor')[['2024 Original Forecast', '2025 Original Forecast']].sum()
cost_type_summary = filtered_data.groupby('Cost Type')[['2024 Original Forecast', '2025 Original Forecast']].sum()

```

```python
import numpy as np
import seaborn as sns

# Calculate the differences between 2024 and 2025 Original Forecasts
filtered_data['Forecast Difference'] = (
    filtered_data['2025 Original Forecast'] - filtered_data['2024 Original Forecast']
)

# Prepare data for Waterfall Chart
waterfall_data = (
    filtered_data.groupby('Department')['Forecast Difference']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

# Waterfall Chart
plt.figure(figsize=(12, 6))
cumulative = waterfall_data['Forecast Difference'].cumsum()
base = np.zeros(len(waterfall_data))
colors = ['green' if val > 0 else 'red' for val in waterfall_data['Forecast Difference']]

for i in range(len(waterfall_data)):
    plt.bar(
        x=waterfall_data['Department'][i],
        height=waterfall_data['Forecast Difference'][i],
        bottom=base[i],
        color=colors[i],
    )
    if i < len(waterfall_data) - 1:
        base[i + 1] = cumulative[i]

plt.title('Waterfall Chart: 2024 vs 2025 Original Forecast Differences by Department')
plt.ylabel('Cost Difference (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Ensure both 2024 and 2025 forecast columns are numeric and clean
filtered_data['2025 Original Forecast'] = (
    filtered_data['2025 Original Forecast']
    .replace('[\$,]', '', regex=True)
    .astype(float, errors='ignore')
)

# Filter out rows with zero or missing values in '2025 Original Forecast'
sunburst_data_cleaned_2025 = filtered_data[filtered_data['2025 Original Forecast'] > 0]

# Recreate Sunburst Diagrams for both years
fig_2024 = px.sunburst(
    sunburst_data_cleaned,
    path=['Department', 'Cost Type'],
    values='2024 Original Forecast',
    color='2024 Original Forecast',
    color_continuous_scale='Viridis',
    title="2024 Original Forecast: Sunburst by Department and Cost Type",
)

fig_2025 = px.sunburst(
    sunburst_data_cleaned_2025,
    path=['Department', 'Cost Type'],
    values='2025 Original Forecast',
    color='2025 Original Forecast',
    color_continuous_scale='Cividis',
    title="2025 Original Forecast: Sunburst by Department and Cost Type",
)

# Combine visualizations into a single HTML
from plotly.subplots import make_subplots

combined_fig = make_subplots(rows=1, cols=2, specs=[[{'type': 'domain'}, {'type': 'domain'}]])
combined_fig.add_trace(fig_2024.data[0], row=1, col=1)
combined_fig.add_trace(fig_2025.data[0], row=1, col=2)

combined_fig.update_layout(
    title_text="Comparison of 2024 and 2025 Original Forecasts: Sunburst by Department and Cost Type",
    annotations=[
        dict(text="2024 Forecast", x=0.2, y=1.1, showarrow=False, font_size=14),
        dict(text="2025 Forecast", x=0.8, y=1.1, showarrow=False, font_size=14),
    ],
)

# Save the combined visualization
combined_file_path = r'C:\Users\dharris\OneDrive - Jazzpharma.com\Project\Jazz.Python\Six Sigma Data with Python\data\sunburst_forecast_comparison_fixed.html'
combined_fig.write_html(combined_file_path)

# Inform the user and provide access to the combined file
combined_file_path
```

```python
# Aggregating forecast data by Department and Functional Area
drilldown_data = filtered_data.groupby(['Department', 'Functional Area'])[['2024 Original Forecast', '2025 Original Forecast']].sum().reset_index()

# Add a column for forecast difference (2025 - 2024) to identify trends/discrepancies
drilldown_data['Forecast Difference'] = drilldown_data['2025 Original Forecast'] - drilldown_data['2024 Original Forecast']

# Sort the data by forecast difference for better visualization
drilldown_data_sorted = drilldown_data.sort_values(by='Forecast Difference', ascending=False)
drilldown_data_sorted

```

```python
# Vendor Analysis: Aggregate cost contributions by Vendor for 2024 and 2025
vendor_analysis = filtered_data.groupby('Vendor')[['2024 Original Forecast', '2025 Original Forecast']].sum().reset_index()

# Add a column for variance (2025 - 2024) to evaluate spending patterns
vendor_analysis['Forecast Difference'] = vendor_analysis['2025 Original Forecast'] - vendor_analysis['2024 Original Forecast']

# Sort by 2024 Original Forecast to rank vendors by cost contribution
vendor_analysis_sorted = vendor_analysis.sort_values(by='2024 Original Forecast', ascending=False)

# Highlight Variances: Filter vendors with the largest absolute differences
top_variances = vendor_analysis_sorted.nlargest(10, 'Forecast Difference')
bottom_variances = vendor_analysis_sorted.nsmallest(10, 'Forecast Difference')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def create_department_comparison(axes, department_costs_percentages):
    # Clear the existing axis
    axes[1, 2].clear()
    
    # Create department comparison
    departments = department_costs_percentages.index
    y_pos = np.arange(len(departments))
    width = 0.35
    
    # Create bars for 2024 and 2025
    bars1 = axes[1, 2].barh(y_pos - width/2, 
                           department_costs_percentages['2024 Original Forecast'], 
                           width, 
                           label='2024', 
                           color='lightblue', 
                           alpha=0.8)
    bars2 = axes[1, 2].barh(y_pos + width/2, 
                           department_costs_percentages['2025 Original Forecast'], 
                           width, 
                           label='2025', 
                           color='orange', 
                           alpha=0.8)

    # Add percentage labels on the bars
    def add_labels(bars):
        for bar in bars:
            width = bar.get_width()
            axes[1, 2].text(width, 
                          bar.get_y() + bar.get_height()/2, 
                          f'{width:.1f}%', 
                          ha='left', 
                          va='center',
                          fontsize=8,
                          padding=5)

    add_labels(bars1)
    add_labels(bars2)

    # Customize the chart
    axes[1, 2].set_yticks(y_pos)
    axes[1, 2].set_yticklabels(departments)
    axes[1, 2].invert_yaxis()  # Departments read top-to-bottom
    axes[1, 2].set_xlabel('Percentage of Total Cost')
    axes[1, 2].set_title('Department Cost Distribution Comparison\n2024 vs 2025')
    axes[1, 2].legend(loc='upper right')
    
    # Add gridlines for better readability
    axes[1, 2].grid(True, axis='x', linestyle='--', alpha=0.7)
    
    # Add a vertical line at 0 for reference
    axes[1, 2].axvline(x=0, color='gray', linestyle='-', alpha=0.3)
    
    # Ensure all labels are visible
    plt.setp(axes[1, 2].get_xticklabels(), rotation=0)
    plt.setp(axes[1, 2].get_yticklabels(), rotation=0)
    
    # Adjust layout to prevent label cutoff
    axes[1, 2].set_xlim(left=0)  # Start at 0%
    right_limit = max(
        department_costs_percentages['2024 Original Forecast'].max(),
        department_costs_percentages['2025 Original Forecast'].max()
    ) * 1.2  # Add 20% padding
    axes[1, 2].set_xlim(right=right_limit)

    return axes[1, 2]

# To use this in your main dashboard code, replace the pie chart section with:
```
